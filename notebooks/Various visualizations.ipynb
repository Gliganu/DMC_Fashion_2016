```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# inputFileName = '..\\data\\orders_train_short.txt'
inputFileName = '..\\data\\orders_train.txt'
data = pd.read_csv(inputFileName, delimiter=';', skipinitialspace=True)
```

Wanted to see how the Return Quantity varied depending on the fact that the item is overpriced or not. It appears it is not helpful

```python
data['overpriced'] = data['price'] > data['rrp']

death_counts = pd.crosstab(data.overpriced, data.returnQuantity.astype(bool))
death_counts.plot(kind='bar', stacked=True, color=['black','gold'], grid=False)
```

```python
data['month']=  data['orderDate'].map(lambda entryDate: float(entryDate.split("-")[1]))
 
 dateCrossTab = pd.crosstab(data.month, data.returnQuantity.astype(bool))
 dateCrossTab.plot(kind='bar', stacked=True, color=['black','gold'], grid=False)
```

```python
data['day']=  data['orderDate'].map(lambda entryDate: float(entryDate.split("-")[2]))
 
 dateCrossTab = pd.crosstab(data.day, data.returnQuantity.astype(bool))
 dateCrossTab.plot(kind='bar', stacked=True, color=['black','gold'], grid=False)
```

```python
dateCrossTab = pd.crosstab(data.deviceID, data.returnQuantity.astype(bool))
 dateCrossTab.plot(kind='bar', stacked=True, color=['black','gold'], grid=False)
```
