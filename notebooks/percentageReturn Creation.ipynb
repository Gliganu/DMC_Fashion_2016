```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# inputFileName = '..\\data\\orders_train_short.txt'
inputFileName = '..\\data\\orders_train_100k.txt'
data = pd.read_csv(inputFileName, delimiter=';', skipinitialspace=True)
```

```python
groupedByCustomer = data['quantity'].groupby(data['customerID'])
```

```python
groupedByCustomer.apply(sum)
```

Now taking the quantity and return quantity and be grouped per customer

```python
byCustomer = data[['quantity','returnQuantity']].groupby(data['customerID'])
```

We sum them grouped by customer. The result is a DF

```python
summedByCustomer = byCustomer.apply(sum)
```

```python
summedByCustomer['percentageReturned'] = summedByCustomer['returnQuantity'] /  summedByCustomer['quantity']
```

```python
summedByCustomer
```

Create the dictionary which will be used later having only id-percentage

```python
finalDataFrame = summedByCustomer.drop(['returnQuantity','quantity'], 1)
idToPercDict = finalDataFrame.to_dict().get('percentageReturned')
idToPercDict
```

Add pe langa coloana de percentageReturn, o coloana cu legatura intre quantity-returnQuantity. boolean ? sa se adauge sum() la fiecare instanta 

```python
data['percentageReturned'] = data['customerID'].apply(lambda id: idToPercDict[id])
```

```python
np.isfinite(data['percentageReturned'].sum()).all()
```
